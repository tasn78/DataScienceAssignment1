Create basic visuals for future use

```python
# Plot histograms and save them
plt.figure(figsize=(12, 8))
for i, column in enumerate(['Height', 'Weight', 'Age', 'Grip strength']):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[column], kde=True, bins=10)
    plt.title(f'Distribution of {column}')

# Save the histogram figure
plt.tight_layout()
plt.savefig('/content/results/histograms.png')  # Save as PNG
```

```python
# Plot boxplots for 'Grip strength' and 'Age' by 'Frailty' and save them
plt.figure(figsize=(12, 6))
sns.boxplot(x='Frality', y='Grip strength', data=frality_df)
plt.title('Grip Strength by Frailty')
plt.savefig('/content/results/grip_strength_boxplot.png')  # Save boxplot image

# Plot another boxplot for 'Age' by 'Frailty' and save it
plt.figure(figsize=(12, 6))
sns.boxplot(x='Frality', y='Age', data=df)
plt.title('Age by Frality')
plt.savefig('/content/results/age_boxplot.png')  # Save boxplot image
```
